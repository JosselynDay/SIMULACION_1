Generador Congruencial 1

```python
import numpy as np
import matplotlib.pyplot as plt
import tabulate

# Parámetros de la congruencia lineal
a = 1664525  # Un valor más grande para mejorar la aleatoriedad
c = 1013904223  # Un valor de desplazamiento comúnmente utilizado en generadores de calidad
m = 2**32  # El valor de m puede ser una potencia de 2, como 2^32, para mejor distribución
xo = 123456789  # Semilla de inicio
n = 102  # Número de iteraciones

# Inicialización de los arreglos
x = np.zeros(n)
xn = np.zeros(n)

def randomnums(a, c, m, xo, n):
    x[0] = xo
    xn[0] = xo / m  # Normalizamos el primer valor

    # Generación de la secuencia de números pseudoaleatorios
    for i in range(1, n):
        x[i] = (a * x[i-1] + c) % m
        xn[i] = x[i] / m

    # Imprimir la tabla de resultados
    encabezados = ["i", "x_i", "x_[i+1]", "x_i / m"]
    tabla = list(zip(range(n), x, np.roll(x, -1), xn))
    print(tabulate.tabulate(tabla, headers=encabezados, tablefmt="fancy_grid"))

    # Graficar los números generados
    plt.figure(figsize=(10, 5))
    plt.plot(range(n), xn, marker='o', linestyle='-', color='b', markersize=4, label='Números Aleatorios')
    plt.xlabel('Índice i')
    plt.ylabel('x_i / m')
    plt.title('NúmerosGenerados')
    plt.legend()
    plt.grid(True)
    plt.show()

randomnums(a, c, m, xo, n)
```
