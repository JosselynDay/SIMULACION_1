```python
import numpy as np
a = 4
c = 7
m = 22
xo = 3
n = 10
x = np.zeros(n)
i=1
x[0]=xo
print("0","         ",int(xo),"          ",xo/m)
while i<n:
  x[i] = ((a *x[i-1]  + c) % m)
  xno = x[i]
  xn = xno/m
  print(i,"         ",int(xno),"          ",xn)
  i=i+1
```

**xn/m**

```python
import numpy as np
import tabulate
a = 4
c = 7
m = 22
xo = 3
n = 50
x = np.zeros(n)
xn = np.zeros(n)

def randomnums(a,c,m,xo,n):
  x[0] = xo
  xn[0] = xo / m
  i = 1
  while i<n:
    x[i] = ((a *x[i-1]  + c) % m)
    xn[i] = x[i]/m
    i=i+1
  encabezados=["i","x_i","x_[i+1]"]
  return(print(tabulate.tabulate(zip(range(i+1),x,xn),headers=encabezados,tablefmt="fancy_grid")))
```

```python
randomnums(a,c,m,xo,n)
```
